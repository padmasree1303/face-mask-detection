```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1/255.0, validation_split=0.2)

train_gen = datagen.flow_from_directory(
    "dataset_raw",
    target_size=(150, 150),
    batch_size=4,
    class_mode="binary",
    subset="training"
)

val_gen = datagen.flow_from_directory(
    "dataset_raw",
    target_size=(150, 150),
    batch_size=4,
    class_mode="binary",
    subset="validation"
)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.summary()
```

```python
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10
)
```

```python
from PIL import Image, ImageDraw
import math

def draw_face(with_mask=True, filename="test_mask.jpg"):
    size = (224, 224)
    bg = (255, 255, 255)
    face = (255, 224, 189)
    eye = (40, 40, 40)
    mask_color = (0, 120, 200)

    img = Image.new("RGB", size, bg)
    draw = ImageDraw.Draw(img)

    # face circle
    cx, cy, r = size[0]//2, size[1]//2 - 10, 70
    draw.ellipse((cx-r, cy-r, cx+r, cy+r), fill=face, outline=(0,0,0))

    # eyes
    ex = 30
    ey = 10
    draw.ellipse((cx-ex-15, cy-ey-10, cx-ex+5, cy-ey+10), fill=eye)  # left
    draw.ellipse((cx+ex-5, cy-ey-10, cx+ex+15, cy-ey+10), fill=eye)  # right

    # nose (triangle)
    nose = [(cx, cy-5), (cx-12, cy+20), (cx+12, cy+20)]
    draw.polygon(nose, fill=(240,180,150))

    if with_mask:
        # mask rectangle with slight curve
        mask_top = cy+20
        mask_bottom = cy+55
        draw.rectangle((cx-60, mask_top, cx+60, mask_bottom), fill=mask_color)
        # mask straps
        draw.line((cx-60, cy+25, cx-95, cy+10), fill=mask_color, width=6)
        draw.line((cx+60, cy+25, cx+95, cy+10), fill=mask_color, width=6)
        # small darker band for fold
        draw.rectangle((cx-60, mask_top+8, cx+60, mask_top+15), fill=(0,90,150))
    else:
        # mouth (smile)
        mouth_box = (cx-30, cy+25, cx+30, cy+45)
        draw.arc(mouth_box, start=200, end=340, fill=(140,0,20), width=4)

    img.save(filename, quality=90)
    print(f"Saved {filename}")

# create both files
draw_face(with_mask=True, filename="test_mask.jpg")
draw_face(with_mask=False, filename="test_no_mask.jpg")
```

```python

from tensorflow.keras.preprocessing import image
import numpy as np

def predict_image_local(path, model, img_size=(224,224)):
    img = image.load_img(path, target_size=img_size)
    arr = image.img_to_array(img) / 255.0
    arr = np.expand_dims(arr, axis=0)
    prob = model.predict(arr)[0][0]
    # NOTE: depending on your training generator, label mapping might be reversed.
    # If your training used directory order ['with_mask','without_mask'] then
    # sigmoid>0.5 likely means class index 1 (without_mask). Adjust as needed.
    label = 'with_mask' if prob <= 0.5 else 'without_mask'
    print(path, "->", label, f"(score={prob:.3f})")

```

```python
# Upload your own image and make prediction

from google.colab import files
from tensorflow.keras.preprocessing import image
import numpy as np

def predict_uploaded_image(model, img_size=(224,224)):
    # 1. Upload image
    uploaded = files.upload()

    for fn in uploaded.keys():
        print("Uploaded:", fn)

        # 2. Load + preprocess
        img = image.load_img(fn, target_size=img_size)
        img_arr = image.img_to_array(img) / 255.0
        img_arr = np.expand_dims(img_arr, axis=0)

        # 3. Predict
        prob = model.predict(img_arr)[0][0]

        # 4. Convert probability → label
        # Adjust if your dataset order is different
        label = "with_mask" if prob <= 0.5 else "without_mask"

        print(f"\nPrediction for {fn}:")
        print("---------------------------------")
        print("Label:", label)
        print("Confidence:", round(float(prob), 4))
        print("---------------------------------")

# Run prediction
predict_uploaded_image(model, img_size=(150,150))   # change to your model size
```
